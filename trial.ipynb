```python
import gradio as gr
import google.generativeai as genai
```

```python
from gemini import GeminiClient
```

```python
client = genai.configure(api_key='add your own')
```

```python
model = genai.GenerativeModel('gemini-1.5-pro')
```

```python
%%time
response = model.generate_content("What is the meaning of life?")
```

```python
for chunk in response:
  print(chunk.text)
  print("_"*80)
```

```python
import PIL.Image

img = PIL.Image.open('ada.jpg')
img
```

```python
response = model.generate_content(["Write a short, engaging blog post based on this woman. She is Ada Lovelace.", img], stream=True)
response.resolve()
```

```python
for chunk in response:
  print(chunk.text)
  print("_"*80)
```
